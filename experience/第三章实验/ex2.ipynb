```python
def dist(tr1,tr2):
    length = len(tr1)
    half = int(length/2)
    x1 = tr1[0:half]
    y1 = tr1[half:]
    x2 = tr2[0:half]
    y2 = tr2[half:]
    length = len(x1)
    deta_x = x1 - x2
    deta_y = y1 - y2
    distance = np.sum((deta_x**2 + deta_y**2) ** 0.5) / length
    distance = distance/10
    print(distance)
    return distance
```

```python
def sample(x,step):
    x_sample = []
    index = np.arange(0,len(x),step)
    for i in index:
        x_sample.append(x[i])
    return np.array(x_sample)
```
